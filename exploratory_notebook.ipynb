```python
# Plot Time Series for all Indicators and the Sub-index

# Variables to plot
variables = ['q1q2_sub-index_oblig', 'q1q2_ind_proced', 'q1q2_ind_actor', 'q1q2_ind_conflict']
countries = q1q2_oblig_df['country'].unique()

# Create a grid of subplots
rows, cols = 7, 6
fig, axs = plt.subplots(nrows=rows, ncols=cols, figsize=(15, 20), sharex=True, sharey=True)

axs = axs.flatten()

# Collect handles for the legend
handles = []
labels = []

for i, country in enumerate(countries):
    country_data = q1q2_oblig_df[q1q2_oblig_df['country'] == country]
    
    for var in variables:
        line, = axs[i].plot(country_data['year'], country_data[var], label=var)
        if i == 0:  # Only need to collect handles and labels from the first subplot
            handles.append(line)
            labels.append(var)
            
    axs[i].set_title(country)

# Add a global legend for the entire figure above the plots
fig.legend(handles, labels, loc='upper center', ncol=len(variables), fontsize=16)#, bbox_to_anchor=(0.5, 1.05))

# Adjust layout to fit everything
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Export figure
plt.savefig("../figures/small-multiples-time-series_obligations_sub-index_indicators.png", format="png", bbox_inches='tight', pad_inches=0.5, dpi=300)

# Display figure
plt.show()
```

```python
#### Ranking of Change
q1q2_oblig_df_sorted = q1q2_oblig_df.sort_values(['country', 'year'])
#q1q2_oblig_df_sorted
```

```python
# Calculate the difference between years (i.e. 2001-2000; 2002-2001; etc. -> first value returns NA)
q1q2_oblig_df_sorted['yearly_change'] = q1q2_oblig_df_sorted.groupby('country')['q1q2_sub-index_oblig'].diff()
# check NA
q1q2_oblig_df_sorted['yearly_change'].isna().sum()
```

```python
# Calculate absolute value (-0.5 -> 0.5; +0.5 -> 0.5)
q1q2_oblig_df_sorted['abs_change'] = q1q2_oblig_df_sorted['yearly_change'].abs()
# check NA
q1q2_oblig_df_sorted['abs_change'].isna().sum()
```

```python
# Calculate sum of all absolute values
reform_intensity = q1q2_oblig_df_sorted.groupby('country')['abs_change'].sum()

# Create dataframe to display Total Reform Score 
reform_intensity = reform_intensity.reset_index()
reform_intensity = reform_intensity.rename(columns={'abs_change': 'total_reform_score'})
reform_intensity = reform_intensity.sort_values('total_reform_score', ascending=False)

# Display ranking
reform_intensity
```

```python
# Bar Chart on Absolute Reform Change per Country

plt.figure(figsize=(15, 10))
plt.bar(reform_intensity['country'], reform_intensity['total_reform_score'])
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Total Reform Intensity')
plt.title('Total Absolute Reform Change by Country')
plt.tight_layout()

# Export figure
plt.savefig("../figures/bar-chart_obligations_absolute-reform-change.png", format="png", bbox_inches='tight', pad_inches=0.5, dpi=300)

# Display figure
plt.show()
```
