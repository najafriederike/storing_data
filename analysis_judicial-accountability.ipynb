```python
# Create overview of country-years with missing values per column (V1)

def overview_nan_df(df):
    
    # Create empty dictionary
    nan_dict = {}
    
    # Iterate through all columns of dataframe:
    for col in df.columns:
        # if column has missing values, create list of country_year where that missing value occurs
        if df[col].isna().any():
            missing_country_years = df.loc[df[col].isna(), 'country_year'].tolist()
           # missing_country_years = ", ".join(map(str, missing_country_years))
            nan_dict[col] = missing_country_years

    return nan_dict
```

```python
df_missing_values = overview_nan_df(q1_q2_df)
df_missing_values
```

```python
# Create overview of country-years with missing values per column (V2 -- Advanced)

def overview_nan_dict_adv(df):
    
    result = {}
    
    for col in df.columns:
        if df[col].isna().any():
            
            missing = df.loc[df[col].isna(), "country_year"]
            
            temp = (pd.DataFrame({"country_year": missing}))
            
            temp["country"] = temp["country_year"].str[:3]
            temp["year"] = temp["country_year"].str[3:].astype(int) + 2000
            
            nested = (
                temp.groupby("country")["year"]
                .apply(list)
                .to_dict()
            )
            
            result[col] = nested
    
    return result
```

```python
# Return overview of country-years with missing values (V1)

df_missing_values_adv = overview_nan_dict_adv(q1_q2_df)

for outer_key, inner_dict in df_missing_values_adv.items():
    print(f"\n*** Question item:{outer_key} ***")
    for inner_key, inner_value in inner_dict.items():
        print(f"  {inner_key}: {inner_value}")
```

```python
# Return overview of country-years with missing values (V2)

df_missing_values_adv = overview_nan_dict_adv(q1_q2_df)

for key, value in df_missing_values_adv.items():
    print(f"\nQuestion item: {key}")
    print(f"Country years: {value}")
```

#### for Check normal distribution
- Check normality with Shapiro-Wilk Test: https://www.geeksforgeeks.org/machine-learning/how-to-perform-a-shapiro-wilk-test-in-python/

```python
# Python's scipy.stats.shapiro

from scipy.stats import shapiro

# Conduct the  Shapiro-Wilk Test
statistic, pvalue = shapiro(q1_q2_df['q2_right'], nan_policy='omit') # 'q1_oblig', 'q1_right', 'q2_oblig', 'q2_right'
print(f"Test statistic: {statistic}")
print(f"p-value: {pvalue}")

# Interpretation
if pvalue > 0.05:
    print("We fail to reject the null hypothesis that the data is normally distributed.")
else:
    print("We reject the null hypothesis that the data is normally distributed.")
```

#### Correlation Matrices

```python
# Correlation coefficient

import scipy.stats
# Spearman Correlation: Works with non-linear data + Suitable for ordinal data -> good fit (?!)
spearman_corr, p_value_s = scipy.stats.spearmanr(q1_q2_df['q1_oblig'], q1_q2_df['q1_right'])
print(f"Spearman's rank correlation coefficient: {spearman_corr}\np-value: {p_value_s}")
print(f"\n")
# Pearson Correlation: assumes normally distributed data -> no good fit
pearson_corr, p_value_p = scipy.stats.pearsonr(q1_q2_df['q1_oblig'], q1_q2_df['q1_right'])
print(f"Pearson's correlation coefficient: {pearson_corr}\np-value: {p_value_p}")
print(f"\n")
# Kendall Correlation: measures rank consistency between variables + more robust for small datasets (?)
kendall_corr, p_value_k = scipy.stats.kendalltau(q1_q2_df['q1_oblig'], q1_q2_df['q1_right'])
print(f"Kendall's rank correlation coefficient: {kendall_corr}\np-value: {p_value_k}")
```
